```python
%%sql
SELECT *
    FROM facts
    LIMIT 5;
```

```python
%%sql

SELECT MIN(population), MAX(population), MIN(population_growth), MAX(population_growth)
    FROM facts;
```

```python
%%sql

SELECT name, population
    FROM facts
    WHERE POPULATION = 0;
```

```python
%%sql

SELECT name, population
    FROM facts
    WHERE POPULATION = 7256490011;
```

```python
%%sql

SELECT MIN(population), MAX(population), MIN(population_growth), MAX(population_growth)
    FROM facts
    WHERE name != "World";
```

```python
%%sql

SELECT AVG(population), AVG(area)
    FROM facts
    WHERE name != 'World';
```

```python
%%sql

SELECT name, population, area
    FROM facts
    WHERE (population > (SELECT AVG(population)
                            FROM facts
                            WHERE name != 'World') AND area < (SELECT AVG(area)
                                                                FROM facts
                                                                WHERE name != 'World'))
```

```python
%%sql

SELECT name, population
    FROM facts
    WHERE population = (SELECT MAX(population)
                            FROM facts
                            WHERE name != 'World')
```

```python
%%sql

SELECT name, population_growth
    FROM facts
    WHERE population_growth = (SELECT MAX(population_growth)
                            FROM facts
                            WHERE name != 'World')
```

```python
%%sql

SELECT name, (area_water/area_land) as ratio_water_to_land
    FROM facts
    ORDER BY ratio_water_to_land DESC
    LIMIT 10;
                            
```

```python
%%sql

SELECT name, area_land, area_water
    FROM facts
    WHERE area_water > area_land;
                     
```

```python
%%sql

SELECT name, (birth_rate - death_rate - migration_rate) as add_people_rate
    FROM facts
    ORDER BY add_people_rate DESC;
```

```python
%%sql

SELECT name, birth_rate, death_rate
    FROM facts
    WHERE death_rate > birth_rate;
```

```python
%%sql

SELECT name, (population/area) as population_area_ratio
    FROM facts
    ORDER BY population_area_ratio DESC;
```
